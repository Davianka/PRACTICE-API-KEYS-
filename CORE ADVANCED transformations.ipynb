```python
super_df[['Hero', 'Publisher']] = super_df['Hero|Publisher'].str.split('|',expand= True)
super_df = super_df.drop(columns = "Hero|Publisher")
super_df
```

```python
super_df['Measurements'] = super_df['Measurements'].str.replace("'",'"')
```

```python
super_df[['Height_cm', 'Weight_kg']] = super_df['Measurements'].str.split(',',expand= True)
super_df = super_df.drop(columns = 'Measurements')
super_df
```

```python
# Make a list of all characters to replace
to_replace = ['{',':', '"','Height', 'cm']
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    super_df['Height_cm'] =super_df['Height_cm'].str.replace(char,'',regex=False)
    
super_df['Height_cm'].head()

```

```python
# Make a list of all characters to replace
to_replace = ['}',':', '"','Weight', 'kg']
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    super_df['Weight_kg'] =super_df['Weight_kg'].str.replace(char,'',regex=False)
    
super_df['Weight_kg'].head()
```

```python
super_df['Weight_kg'] = super_df['Weight_kg'].astype(float)
super_df['Height_cm'] = super_df['Height_cm'].astype(float)
super_df.info()
```
